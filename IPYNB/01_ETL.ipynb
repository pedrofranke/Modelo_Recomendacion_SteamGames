## Desanidacion de columnas

```python
# verifiquemos cuantas filas deberian haber en items, aprovechando que tiene la cantidad de registros por fila anidada:

registros = dic['users_items']['items_count'].sum()
```

```python
columnas = {'user_reviews':'reviews','users_items':'items'} #las columnas a desanidar

for k,v in columnas.items(): #iteramos sobre el diccionario
    dic[k] = dic[k].explode(v) #expandimos la lista de la columna anidada para poder leer el json
    dic[k].dropna(subset=v,inplace=True) #eliminamos los registros que no tienen informacion
    dic[k] = pd.json_normalize(dic[k].to_dict(orient='records'), sep='') #normalizamos la columna, desanidandola
    dic[k] = dic[k].rename(columns=lambda x: x.replace(v, '')) #eliminamos el prefijo del nombre de las columnas
```

```python
filas = dic['users_items'].shape[0] #contamos la cantidad de registros para comparar

print(filas,registros,filas == registros) #comparamos los registros indicados en las columnas vs los obtenidos, no perdimos informacion
```

### Seleccion de columnas

```python
columns = ['genres','release_date','price','id','developer','app_name']

df = df[columns]
```

### Tipo de datos

```python
md.tipo_datos(df)
```

#### Columnas Publisher, app_name, developer

Completaremos los nulos con 'Sin_Dato'

```python
columnas = ['app_name','developer']

df[columnas] = df[columnas].fillna('Sin_Dato')

md.tipo_datos(df)
```

#### Columna Price

Corrijiremos los valores str de la columna price.

```python
df.price.unique()
```

```python
# enviamos todos los textos a float valor 0.0 si es que no aclaran nada en el texto de valores
df['price'] = df['price'].apply(md.col_price)
df.price.unique()
```

#### Columna release_date

Corrigiremos los valores str y solo tomaremos la informacion del anio de lanzamiento, o 'Sin_Dato' si es que no ha sido lanzado.

```python
df['release_year'] = df['release_date'].apply(md.release)
df['release_year'].unique()
```

```python
df.drop(columns='release_date',inplace=True) # drop a la columna que ya no nos sirve mas
```

```python
df
```

#### Columna Genres

Apliaremos la columna para tener una base en donde exista una linea por cada genero, repitiendose el juego.

```python
df = df.explode('genres')
df.dropna(subset=['genres'],inplace=True)
df
```

#### Estandarizacion de tipeo

```python
text_cols = ['genres','developer','app_name']

df[text_cols] = df[text_cols].apply(md.convertir_proper)
```

```python
dic['steam_games'] = df.copy()
```
